```python
# First cell: Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Read Data Using Pandas
df = pd.read_csv('banking.csv')
df.head()
```

```python
# 2. Exploring Data Frames
print("Data shape:", df.shape)
print("\nData types:")
print(df.dtypes)
print("\nSummary statistics:")
df.describe()
```

```python
#3. Data Frame Data Types
print("Current dtypes:")
print(df.dtypes)

# Convert types example - using an existing column
df['job'] = df['job'].astype('category')
print("\nAfter conversion:")
print(df.dtypes)
```

```python
#4. Data Frames Attributes
print("Columns:", df.columns)
print("\nAxes:", df.axes)
print("\nNumber of dimensions:", df.ndim)
print("\nSize:", df.size)
print("\nShape:", df.shape)
```

```python
# 5. Data Frame Methods
print("First 3 rows:")
print(df.head(3))
print("\nStatistics:")
print(df.describe())

# Option 1: Get max values for numeric columns only (recommended)
print("\nMax values (numeric columns only):")
print(df.select_dtypes(include=['number']).max())

# Option 2: Alternatively, you can specify numeric_only=True
# print("\nMax values (numeric columns only):")
# print(df.max(numeric_only=True))

print("\nRandom sample:")
print(df.sample(3))
```

```python
# 6. Select a Column in a Data Frame
print("Single column:")
print(df['age'].head())  # Using 'age' which exists in your DataFrame
print("\nMultiple columns:")
print(df[['age', 'job']].head())  # Using existing columns 'age' and 'job'
```

```python
# 7. Data Frame GroupBy Method
# Groupby example
grouped = df.groupby('age')
print("Mean by age:")
# Convert categorical columns to numeric before calculating mean
print(grouped.mean(numeric_only=True))
```

```python
# 8. Data Frame: Filtering (Bank Marketing Data)
print("Filter age > 30:")
print(df[df['age'] > 30].head())  # ✅ Correct column name

print("\nFilter for clients with housing loans:")
print(df[df['housing'] == 'yes'].head())  # ✅ Correct column name
```

```python
# 9. Data Frame: Slicing
print("Rows 5-10:")
print(df.iloc[5:10])
print("\nSpecific rows and columns:")
print(df.iloc[5:10, [0, 4]])
```

```python
# 10. Data Frame: Selecting Rows
print("Row with index 3:")
print(df.iloc[3])
print("\nRows with specific indices:")
print(df.iloc[[1, 5, 10]])
```

```python
# 11. Data Frame: Method Loc (Bank Dataset)
print("Row with index 3 using loc:")
print(df.loc[3])  # Correct (shows row 3)

print("\nFilter and select columns:")
# Example: Filter married clients and show 'age' and 'marital' status
print(df.loc[df['marital'] == 'married', ['age', 'marital']].head())
```

```python
# 12. Data Frame: Sorting (Bank Dataset)
print("Sorted by client age:")
print(df.sort_values('age').head())  # Youngest to oldest clients

print("\nSorted descending by call duration:")
print(df.sort_values('duration', ascending=False).head())  # Longest calls first
```

```python
# 13. Aggregation Function in Pandas (Bank Data)
print("Custom aggregations for bank data:")
print(df.agg({
    'age': ['min', 'max', 'mean'],        # Client age statistics
    'duration': ['median', 'std'],        # Call duration statistics
    'campaign': ['mean', 'max']           # Marketing campaign contacts
}))
```

```python
# 14. Descriptive Statistics (Bank Data)
print("Skewness for numerical features:")
print(df.select_dtypes(include='number').skew())

print("\nKurtosis for numerical features:")
print(df.select_dtypes(include='number').kurt())
```

```python
#15. Graphics

import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
df = sns.load_dataset('iris')

plt.figure(figsize=(15, 10))

# Histogram
plt.subplot(2, 3, 1)
sns.histplot(data=df, x='sepal_length')
plt.title('Histogram')

# Boxplot
plt.subplot(2, 3, 2)
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Boxplot')

# Violin plot
plt.subplot(2, 3, 3)
sns.violinplot(data=df, x='species', y='petal_length')
plt.title('Violin plot')

# Scatter plot
plt.subplot(2, 3, 4)
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Scatter plot')

# Pair plot (outside subplot grid)
sns.pairplot(data=df, hue='species')
plt.suptitle('Pair plot', y=1.02)

plt.tight_layout()
plt.show()
```
