```python
# Step 1: Read MathStu.csv data Using pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("Math-Students.csv")
```

```python
# Step 2: Find the Data types of every attribute
print("Data Types:\n", df.dtypes)
```

```python
# Step 3: Describe the statistics for each numeric column
print("\nStatistics:\n", df.describe())
```

```python
# Step 4: Find the Standard Deviation of age
print("\nStandard Deviation of Age:", df["age"].std())
```

```python
# Step 5: Select first 10 data from “famsize” column
print("\nFirst 10 'famsize' values:\n", df["famsize"].head(10))
```

```python
# Step 6: Find mean, median and Std Dev for “freetime”
print("\n'freetime' - Mean:", df["freetime"].mean())
print("'freetime' - Median:", df["freetime"].median())
print("'freetime' - Std Dev:", df["freetime"].std())
```

```python
# Step 7: Calculate mean age for each “address”
print("\nMean age by address:\n", df.groupby("address")["age"].mean())
```

```python
# Step 8: Select only rows whose Mother’s job (Mjob) is teacher
teachers = df[df["Mjob"] == "teacher"]
print("\nRows with Mjob as 'teacher':\n", teachers)
```

```python
# Step 9: Select the rows where age > 15
older_students = df[df["age"] > 15]
print("\nRows with age > 15:\n", older_students)
```

```python
# Step 10: Create new data sorted by Age
sorted_by_age = df.sort_values(by="age")
print("\nData sorted by age:\n", sorted_by_age)
```

```python
# Step 11: Find Min, Max, and Mean of age and freetime
print("\nMin Age:", df["age"].min())
print("Max Age:", df["age"].max())
print("Mean Age:", df["age"].mean())
```

```python
print("\nMin Freetime:", df["freetime"].min())
print("Max Freetime:", df["freetime"].max())
print("Mean Freetime:", df["freetime"].mean())
```

```python
# Step 12: Draw Histogram for age data
plt.figure(figsize=(8, 5))
plt.hist(df["age"], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Step 13: Display Bar plot for Mjob and average age
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Mjob", y="age", ci=None, estimator='mean')
plt.title("Average Age per Mother's Job")
plt.xlabel("Mother's Job")
plt.ylabel("Average Age")
plt.xticks(rotation=45)
plt.show()

```

```python
# Step 14: Create Box plot for famsize and freetime
plt.figure(figsize=(8, 5))
sns.boxplot(x="famsize", y="freetime", data=df)
plt.title("Box Plot of Freetime by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Free Time")
plt.show()
```
